```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("strassen.txt", sep = ',')
first = data[0:5]
second = data[data["Num"] == 7]
second = data[5:10]
```

```python
data
```

```python
import matplotlib
matplotlib.rcParams['lines.linewidth'] = 5
matplotlib.rcParams['lines.linestyle'] = '--'
matplotlib.rcParams['figure.figsize'] = (15,15)
matplotlib.rcParams['legend.fontsize'] = 30
matplotlib.rcParams['font.size'] = 30
```

```python
# plt.figure(figsize=(15, 10))
plt.plot(first["Num"], first["P"], label = 'parallel')
plt.plot(first["Num"], first["NP"], label = 'non-parallel')
plt.legend()
plt.grid()
plt.title('N = 1024')
plt.ylabel('time, s')
plt.xlabel('thread num')
```

```python
plt.figure(figsize=(15, 10))
plt.plot(second["N"], second["P"], label = 'parallel')
plt.plot(second["N"], second["NP"], label = 'non-parallel')
plt.legend()
plt.grid()
plt.title("Thread num = 7")
plt.ylabel('time, s')
plt.xlabel('N')
```

```python
plt.figure(figsize=(15, 10))
plt.plot(second["N"], second["NP"]/second["P"])
plt.grid()
plt.title("Thread num = 7")
plt.ylabel('acceleration, s')
plt.xlabel('N')
```
